```python
#ms
insert_25 = [2333, 2401, 3809, 2866, 4040]
update_25 = [3809, 2984, 2676, 2645, 2633]

insert_500 = [64472, 57259, 54323, 68232, 60102]
update_500 = [55459, 62332, 59912, 65023, 62302]

read_25 = [889, 601, 362, 334, 358]
read_500 = [16213, 15802, 14328, 17391, 16202]

get_v = [959, 1017, 1051, 1346, 1609]
```

```python
import matplotlib.pyplot as plt
import numpy as np
def plot_bar(data,name, xlabel=None, ylabel=None, title=None):
    mean = np.mean(data, axis=1)
    std = np.std(data, axis=1)
    plt.bar(name, mean, yerr=std)

    if ylabel:
        plt.ylabel(ylabel)
    if xlabel:
        plt.xlabel(xlabel)
    if title:
        plt.title(title)
    plt.setp(plt.gca().get_xticklabels(), rotation=25, horizontalalignment='right')
```

```python
plot_bar(
    [insert_25, read_25, update_25, insert_500, read_500, update_500], 
    ['insert_25MB', 'read_25MB', 'update_25MB', 'insert_500MB', 'read_500MB', 'update_500MB'], 
    ylabel="time (ms)",
    title="Insert, Read, Update Time"
)
```

```python
plt.ylabel("time (ms)")
plt.xlabel("num-version")
plt.title("Get Version vs. Time")
plt.errorbar(range(1,6), get_v, yerr=[23, 40, 50, 33, 23], fmt='-o')
```
